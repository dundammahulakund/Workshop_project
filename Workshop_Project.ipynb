```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

demo = data.groupby("year")["amount"].sum()
data

plt.plot(demo.index, demo.values,marker="o",markerfacecolor="red",color="black",markeredgecolor="green",linestyle="--",linewidth=1.5,label='Total')
plt.title("year wise Amount investment",color="red",fontsize=14)
plt.grid(linestyle=":",color="skyblue",linewidth=1.5)
plt.xlabel("year",color="blue",fontsize=12)
plt.ylabel("Total Amount",color="blue",fontsize=12)
plt.legend()
plt.show()
```

```python
Demo = data["city"].value_counts().head(10)
color=["black","red","yellow","pink","green","purple","orange","brown","skyblue","blue"]
plt.bar(Demo.index,Demo.values,color=color,edgecolor="black")


plt.xticks(rotation=45)
plt.title("Top 10 startup company's in India")
plt.xlabel("City Names")
plt.ylabel("Number of start up")
plt.grid(True)
```

```python
# 3. Which startup sectors have the highest number of funded startups?
Demo = data["vertical"].value_counts().head(10)
color=["skyblue","pink","yellow","gray","brown","darkorange","black","lightgreen","purple","lightblue"]
plt.figure(figsize=(10, 6))
plt.barh(Demo.index[::-1],Demo.values[::-1],color=color,alpha=0.6,edgecolor="black")
plt.title("Top 10 Sectors startup in india")
plt.xlabel("Number of startup company")
plt.ylabel("Sectors")
plt.grid(axis='x',linestyle="--",color="black")
```

```python
#4. what is the distributtion of different funding rounds?
Demo = data["round"].value_counts().head(7)
plt.figure(figsize=(10,5))
color = ["blue","red","skyblue","pink","orange","grey","green"]
plt.pie(Demo.values,labels = Demo.index,autopct="%1.1f%%")
plt.title("Rounds distribution in percentage")
```

```python
#5. how are startup funding amounts distribution?

Demo = data["amount"].value_counts().head(7)

plt.hist(data["amount"],edgecolor="black")
```

```python
#6. The Relationship between Funding Year and Funding Amount?
Demo = data["year"].value_counts()
plt.scatter(data["year"],data["amount"],color="skyblue")
plt.yticks()
plt.title("Funding amount by year")
plt.xlabel("year")
plt.ylabel("Amount")
plt.grid(linestyle="--")
plt.show()
```

```python
#7. Which funding round occurs most frequently?
import seaborn as sns

Demo = data["round"].value_counts().head(10).index

sns.countplot(data,y="round",order = Demo,palette="viridis")
plt.title("Most frequently foundin round")
```
